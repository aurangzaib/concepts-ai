```python
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np
```

```python
# Dataset
n_samples = 20
samples = np.random.multivariate_normal(
    mean=[0, 3], cov=[[1.0, 0.5], [0.5, 1.0]], size=n_samples
).astype(np.float32)
targets = np.ones(shape=(n_samples, 1)).astype(np.float32)

# Shuffle
indices = np.random.permutation(n_samples)
samples = samples[indices]
targets = targets[indices]

# Train/Val/Test
n_test, n_val = int(0.1 * n_samples), int(0.2 * n_samples)
samples_train = samples[n_test + n_val :]  # 70%
samples_val = samples[n_test : n_test + n_val]  # 20%
samples_test = samples[:n_test]  # 10%
targets_train = np.ones(shape=(len(samples_train), 1))
targets_val = np.ones(shape=(len(samples_val), 1))
```

```python
model = tf.keras.Sequential(tf.keras.layers.Dense(units=1))
model.compile(
    loss=tf.keras.losses.MeanSquaredError(),
    optimizer=tf.keras.optimizers.RMSprop(learning_rate=0.1),
    metrics=[tf.keras.metrics.BinaryAccuracy()],
)
history = model.fit(
    x=samples_train,
    y=targets_train,
    epochs=100,
    validation_data=(samples_val, targets_val),
)

print(history.history)
targets_test = model.predict(x=samples_test)
```
